```python
# This script takes the fridge files inside of 'original_fridge_files' and reads each one to 
# find when the fridge is actually using significant amounts of power, then places that data
# into its own file inside of a folder called 'fridge_files'. To adjust what is considered as
# "significant amounts of power", adjust 'LOWER_LIMIT' below.


def separator(file):

    LOWER_LIMIT = 15
    lines = []
    
    with open(file) as f:
        line = f.readline()
        line = f.readline()
        while line:
            #print(line)
            if line[-1] == '\n':
                 line = line[:-1]
            line_vals = line.split(",")
            #print(line_vals[2])
            if(float(line_vals[2]) > LOWER_LIMIT):
                lines.append(line)
            elif (float(line_vals[2]) <= LOWER_LIMIT) and (len(lines) > 0):
                if(not os.path.exists(os.getcwd() + '/fridge_files')):
                    os.makedirs(os.getcwd() + '/fridge_files')

                path, dirs, files = next(os.walk(os.getcwd()+ '/fridge_files'))
                with open(os.getcwd() + '/fridge_files/' + 'fridge_' + str(len(files)+1) + '.csv', 'w', newline='\n') as f1:
                    writer = csv.writer(f1)
                    writer.writerow(['power_factor','phase_angle','power_real','power_reactive','power_apparent','Vrms','Irms'])

                    for i in range(len(lines)):
                        vals = lines[i].split(",")
                        #print(vals)
                        writer.writerow(vals)
                lines = []

            line = f.readline()
            
path, dirs, files = next(os.walk(os.getcwd()+ '/original_fridge_files'))   
for i in files:
    separator(os.getcwd()+ '/original_fridge_files/' + i)
```

```python
###############################  THIS IS THE MAIN SCRIPT #############################################

# Currently the script grabs 'NUMBER_OF_TRIALS' files and does superpositon
# with all of them to simulate one whole day of a house. If one of the appliances
# is periodical, it will only add 'APPLIANCE_DURATION' seconds of that specific appliance.
# If it is multi pattern, it will add the entire file. The data is collected from the 'data'
# folder and once all the samples have been collected, a file is generated in the 'aggregate'
# folder. The aggregate file has the total power signal as the first column and an integer that
# represents one of 33 different scinarios that are possible with the five appliances we are using.
# All of the combinations are present in the 'dictionary.txt' file with their corresponding vallue.
# If we decide to add another appliance, we can simply run the script above to generate all of the combinations.

# THE SCRIPT IS CURRENTLY SET UP SUCH THAT EACH APPLIANCE WILL BE ON AT MOST ONCE AT ANY GIVEN SECOND, I.E.
# A LAMP CANNOT BE ON TWICE AT 12:05:15. 

# NEXT STEPS:
# Finding adequite values for 'APPLIANCE_DURATION'


def synthetic_data():

    #Constant
    SECONDS_IN_A_DAY: int = 86400
    NUMBER_OF_TRIALS: int = random.randint(15,40)
    directory = os.getcwd()+'/'
    APPLIANCE_TYPE = {
        # PERIODICAL APPLIANCES
        "television": "periodical",
        "lamp": "periodical",
        # MULTI PATTERN APPLIANCES
        "fridge": "multi",
        "coffee": "multi"

    }    


    # APPLIANCE_LABELS = { 

    #     "lamp": 1,
    #     "fridge" : 2,
    #     "hairdryer" : 3,
    #     "television" : 4,
    #     "coffee" : 5

    # } 


    APPLIANCE_COUNTER = { #currently only multi pattern appliances, can add periodical too if needed
        # MULTI PATTERN APPLIANCES
        "coffee": 25,
        "fridge": 25

    } 
    
    APPLIANCE_DURATION_MIN = {
        # ONLY PERIODICAL APPLIANCES
        "television": 300,
        "lamp": 60,
        "hairdryer": 30
    } 

    APPLIANCE_DURATION_MAX = {
        # ONLY PERIODICAL APPLIANCES
        "television": 21600,
        "lamp": 10800,
        "hairdryer": 300
    }    

    #FOR ANYONE WITH A MAC. IF YOU DONT HAVE A MAC AND ARE GETTING AN ERROR BECAUSE OF THIS LINE, COMMENT IT OUT
    !find . -name ".DS_Store" -print -delete

    #Fill this array with superposition data of various appliances    
    data = np.zeros((SECONDS_IN_A_DAY), dtype=None)
    appliances = np.zeros((SECONDS_IN_A_DAY), dtype='object')
    for i in range(appliances.shape[0]):
        appliances[i] = []
    #print(data)


    # Setting up the dictionary for the various options
    appliance_dict = {}
    with open("dictionary.txt") as f:
        line = f.readline()
        while line:
            #print(line)
            if line[-1] == '\n':
                line = line[:-1]
            temp = line.rsplit(" ",1)
            appliance_dict[temp[0]] = int(temp[1])
            line = f.readline()


    for i in range(NUMBER_OF_TRIALS):
        appliance_type = ""
        current_file = ""
        cur_file_path = ""
        while appliance_type == "":
            path1, dirs1, files1 = next(os.walk(directory+ 'data/'))
            dir_num = random.randint(0,len(dirs1)-1)
            current_file = random.choice(os.listdir(directory + 'data/' + dirs1[dir_num] ))
            cur_file_path = directory + 'data/' + dirs1[dir_num]
            partitioned_string = current_file.partition('_')
            appliance_type = partitioned_string[0]
            #print(appliance_type)
            if APPLIANCE_TYPE[appliance_type] == 'multi' and APPLIANCE_COUNTER[appliance_type] == 0:
                appliance_type = ""
            else:
                if APPLIANCE_TYPE[appliance_type] == 'multi':
                    APPLIANCE_COUNTER[appliance_type] = APPLIANCE_COUNTER[appliance_type] - 1
                
        #taking the appliance that was selected and adding it to the aggregate
        with open(directory + 'data/' + dirs1[dir_num] + '/' + current_file,'r') as csvfile:
            file_elements = csv.reader(csvfile, delimiter = ',')

            iterVals = iter(file_elements)
            x = []
            if(APPLIANCE_TYPE[appliance_type] == 'periodical'):
                l = list(iterVals)
                iterSize = len(l)
                a_size = random.randint(APPLIANCE_DURATION_MIN[appliance_type], APPLIANCE_DURATION_MAX[appliance_type])
                iterStart = 1
                if(iterSize-a_size > 2):
                    iterStart = random.randint(1,iterSize-a_size-1)
                
                #print(a_size)
                while(a_size > 0):
                    
                    for row in range(iterSize - iterStart - 1):
                        x.append(float(l[iterStart+row][2]))
                        a_size = a_size - 1
                        if a_size == 0:
                            break
                    current_file = random.choice(os.listdir(directory + 'data/' + dirs1[dir_num] ))
                    with open(directory + 'data/' + dirs1[dir_num] + '/' + current_file,'r') as csvfile1:
                        file_elements = csv.reader(csvfile1, delimiter = ',')
                        iterVals = iter(file_elements)
                        l = list(iterVals)
                        iterSize = len(l)
                        iterStart = 1
                        if(iterSize-a_size > 2):
                            iterStart = random.randint(1,iterSize-a_size-1)
                    
#                 for row in range(APPLIANCE_DURATION[appliance_type]):
#                     x.append(float(l[iterStart+row][2]))
            else:
                next(iterVals)
                for row in iterVals:
                    x.append(float(row[2]))

            starting_index = random.randint(0,len(data)-len(x))
            #print(starting_index)
            #temp_list = [0 for i in range(starting_index)] + x + [0 for i in range(starting_index+len(x)-1,len(data)-1)]
            temp_list = np.zeros((SECONDS_IN_A_DAY), dtype=None)
            x = np.array(x)
            #putting the appliance in the appliances list
            for j in range(starting_index,starting_index+len(x)):
                if appliance_type in appliances[j]:
                    continue
                else:
                    temp_list[j] = x[j-starting_index]
                    appliances[j].append(appliance_type)
            A1 = np.array(temp_list)
            data = np.array(A1+data)

    #adding noise to elements that were previously 0
    noiselist = []
    with open(directory + 'noise/' + random.choice(os.listdir(directory + 'noise')), 'r') as csvfile:
        file_elements = csv.reader(csvfile, delimiter = ',')

        iterVals = iter(file_elements)
        next(iterVals)
        for row in iterVals:
            noiselist.append(float(row[2]))

    noiselist=np.array(noiselist)

    for i in range(len(data)):
        if data[i] == 0:
            data[i] = noiselist[random.randint(0,len(noiselist)-1)]

    #plotting
#     plt.plot(data)
#     plt.show()


    #Creating aggregate CSV file with first column as total power and second column a list of the appliances on at that time
    #Placing all the files inside an /aggregate directory
    if(not os.path.exists(directory + '/aggregate')):
        os.makedirs(directory + '/aggregate')

    path, dirs, files = next(os.walk(directory+ 'aggregate/'))

    with open(directory+ 'aggregate/' +'aggregate_'+str(len(files)+1)+'.csv', 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Power (W)', 'Appliance List'])
        for i in range(SECONDS_IN_A_DAY):
            list_out = str(sorted(appliances[i]))[1:-1]
            if list_out == '':
                list_out = 'noise'
            writer.writerow([data[i],appliance_dict[list_out]])
```

```python
# This script takes the fridge files inside of 'original_fridge_files' and reads each one to 
# find when the fridge is actually using significant amounts of power, then places that data
# into its own file inside of a folder called 'fridge_files'. To adjust what is considered as
# "significant amounts of power", adjust 'LOWER_LIMIT' below.


def separator_coffee(file):

    LOWER_LIMIT = 50
    lines = []
    
    with open(file) as f:
        line = f.readline()
        line = f.readline()
        while line:
            #print(line)
            if line[-1] == '\n':
                 line = line[:-1]
            line_vals = line.split(",")
            #print(line_vals[2])
            if(float(line_vals[2]) > LOWER_LIMIT):
                lines.append(line)
            elif (float(line_vals[2]) <= LOWER_LIMIT) and (len(lines) > 0):
                if(not os.path.exists(os.getcwd() + '/coffee_files')):
                    os.makedirs(os.getcwd() + '/coffee_files')

                path, dirs, files = next(os.walk(os.getcwd()+ '/coffee_files'))
                with open(os.getcwd() + '/coffee_files/' + 'coffee_' + str(len(files)+1) + '.csv', 'w', newline='\n') as f1:
                    writer = csv.writer(f1)
                    writer.writerow(['power_factor','phase_angle','power_real','power_reactive','power_apparent','Vrms','Irms'])

                    for i in range(len(lines)):
                        vals = lines[i].split(",")
                        #print(vals)
                        writer.writerow(vals)
                lines = []

            line = f.readline()
            
path, dirs, files = next(os.walk(os.getcwd()+ '/original_coffee_files'))   
for i in files:
    separator_coffee(os.getcwd()+ '/original_coffee_files/' + i)
```
